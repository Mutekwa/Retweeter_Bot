```python
import random
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get("https://twitter.com/i/flow/login")

time.sleep(15)

# enter username
username = driver.find_element(By.TAG_NAME, "input")
username.send_keys("AllAdamsAll")


all_buttons = driver.find_elements(
    By.XPATH, "//div[@role='button']"
)
all_buttons[-2].click()

time.sleep(2)

# enter password
password = driver.find_element(
    By.XPATH, "//input[@type='password']"
)
password.send_keys("All@dAms")


time.sleep(2)

#Click the log in button
all_buttons = driver.find_elements(
    By.XPATH, "//div[@role='button']"
)
all_buttons[-1].click()

time.sleep(2)

#search for keyword
keyword = "python"
driver.get("https://twitter.com/search?q=" + keyword + "&src=typed_query")

time.sleep(5)

#list of random messages 
messages = ["Python for life","PyLove","Wonderful","Part of Python"]

#number of scrolls before retweeting
n_scrolls = 3

for scroll in range(n_scrolls):
    # click retweet
    retweet = driver.find_elements(
        By.XPATH, "//div[@data-testid='retweet']"
    )
    driver.execute_script("arguments[0].click();",retweet[-1])

    time.sleep(4)

    # click quote tweet
    quote_tweet = driver.find_element(
        By.XPATH, "//a[@role='menuitem']"
    )
    quote_tweet.click()

    time.sleep(2)
    
    # to quote the tweet from random selection of list messages
    quote = driver.find_element(By.XPATH, "//div[contains(@class, 'public-DraftStyleDefault-block')]")
    quote.send_keys(random.choice(messages))

    time.sleep(2)

    # send the retweet 
    tweet = driver.find_element(
    By.XPATH, "//div[@data-testid='tweetButton']"
    )
    tweet.click()
    
    time.sleep(5)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
```
